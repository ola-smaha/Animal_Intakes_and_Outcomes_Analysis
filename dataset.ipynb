Sonoma data (26.3K rows)<br>
Data Last Updated:
June 14, 2023<br>
Metadata Last Updated:
April 4, 2023

```python
url = "https://data.sonomacounty.ca.gov/resource/924a-vesw.json"

# Initialize an empty list to store the data
data_list = []

# Make requests until all data is retrieved
page = 1
while True:
    # Set the parameters for the page and limit (number of records per page)
    params = {"$limit": 1000, "$offset": (page - 1) * 1000}
    
    # Make the API request
    response = requests.get(url, params=params)
    
    # Check if there is more data
    if response.status_code == 200 and len(response.json()) > 0:
        data_list.extend(response.json())
        page += 1
    else:
        break

# Convert the data to a DataFrame
df = pd.DataFrame(data_list)
```
